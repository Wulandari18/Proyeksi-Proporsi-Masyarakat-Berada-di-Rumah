# Perform Feature Standerd Scalling

```python
all_columns = dataset_sel.columns.tolist()
print(f'All Columns : {all_columns} \n\n')
```

```python
cut_dataset = dataset_sel[['polygon_name_Bandung', 'polygon_name_Bandung Barat', 'polygon_name_Banjar', 'polygon_name_Bekasi', 
                       'polygon_name_Bogor', 'polygon_name_Ciamis', 'polygon_name_Cianjur', 'polygon_name_Cimahi', 
                       'polygon_name_Cirebon', 'polygon_name_Depok', 'polygon_name_Garut', 'polygon_name_Indramayu', 
                       'polygon_name_Karawang', 'polygon_name_Kota Bandung', 'polygon_name_Kota Bekasi', 
                       'polygon_name_Kota Bogor', 'polygon_name_Kota Cirebon', 'polygon_name_Kota Sukabumi', 
                       'polygon_name_Kota Tasikmalaya', 'polygon_name_Kuningan', 'polygon_name_Majalengka', 
                       'polygon_name_Purwakarta', 'polygon_name_Subang', 'polygon_name_Sukabumi', 'polygon_name_Sumedang', 
                       'polygon_name_Tasikmalaya']]
```

```python
cut_dataset.head()
```

```python
y = dataset_sel['ratio_single_tile_users']
X = dataset_sel.drop(['ratio_single_tile_users', 'polygon_name_Bandung',
                      'polygon_name_Bandung Barat', 'polygon_name_Banjar', 'polygon_name_Bekasi', 
                       'polygon_name_Bogor', 'polygon_name_Ciamis', 'polygon_name_Cianjur', 'polygon_name_Cimahi', 
                       'polygon_name_Cirebon', 'polygon_name_Depok', 'polygon_name_Garut', 'polygon_name_Indramayu', 
                       'polygon_name_Karawang', 'polygon_name_Kota Bandung', 'polygon_name_Kota Bekasi', 
                       'polygon_name_Kota Bogor', 'polygon_name_Kota Cirebon', 'polygon_name_Kota Sukabumi', 
                       'polygon_name_Kota Tasikmalaya', 'polygon_name_Kuningan', 'polygon_name_Majalengka', 
                       'polygon_name_Purwakarta', 'polygon_name_Subang', 'polygon_name_Sukabumi', 'polygon_name_Sumedang', 
                       'polygon_name_Tasikmalaya'], axis = 1)
cols = X.columns
index = X.index
```

```python
X.head()    
```

```python
# Scale the data to keep all the values in the same magnitude of 0 -1 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X= sc.fit_transform(X)
X = pd.DataFrame(X, columns = cols, index = index)
```

```python
X = pd.concat([X,cut_dataset], axis = 1)
```

```python
X.head()
```

# Model Implementation

### Train Test Splitting

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=21)
```

```python
X_train
```

```python
X_test
```

### Model Selection

#### Import Machine Learning

```python
from sklearn import linear_model

from sklearn.tree import DecisionTreeRegressor

from sklearn.neighbors import KNeighborsRegressor
```

#### Import Machine Learning Libraries

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
```

# Decision Tree Regressor

### Tunning Parameters

```python
from  sklearn.model_selection import RandomizedSearchCV
```

```python
model = DecisionTreeRegressor()

param_grid = {'max_depth' : np.arange(4, 20), 'min_samples_leaf' : np.arange(3, 20)}
dtc_model = RandomizedSearchCV(model, param_grid, cv = 5)
dtc_model.fit(X_train, y_train)

print("\n Best Score is ")
print(dtc_model.best_score_)

print("\n Best Estimator is ")
print(dtc_model.best_estimator_)

print("\n Best Parametes are")
print(dtc_model.best_params_)
```

```python
y_pred = dtc_model.predict(X_test)
df_pred = pd.DataFrame(y_pred)
```

```python
df_pred = pd.DataFrame(data=[],columns=["y_pred"])
df_pred["y_pred"] = y_pred
df_pred.index = df_pred.index
df_pred.head()
```

```python
mse = mean_squared_error(y_test, y_pred)
print('Mean Squared Error of Decision Tree Regression is:', mse)
```

### Visualization

```python
plt.hist(y_pred, bins = 'sqrt', histtype = 'step', label = 'Predicted')
plt.hist(y_test, bins = 'sqrt', histtype = 'step', label = 'Actual')
plt.legend(loc = 'upper right')
plt.xlabel('Proporsi Masyarakat Berada di Rumah')
plt.savefig('DTC.png', dpi = 300)

plt.show()
```

# K-Nearest Neighbors Regressor

### Tunning Parameters

```python
model = KNeighborsRegressor()

param_grid = {'n_neighbors' : np.arange(2, 50), 'weights' : ['distance', 'uniform']}
knn_model = RandomizedSearchCV(model, param_grid, cv = 10)
knn_model.fit(X_train, y_train)

print("\n Best Score is ")
print(knn_model.best_score_)

print("\n Best Estimator is ")
print(knn_model.best_estimator_)

print("\n Best Parametes are")
print(knn_model.best_params_)
```

```python
y_pred = knn_model.predict(X_test)
df_pred = pd.DataFrame(y_pred)
```

```python
df_pred = pd.DataFrame(data=[],columns=["y_pred"])
df_pred["y_pred"] = y_pred
df_pred.index = df_pred.index
df_pred.head()
```

```python
mse = mean_squared_error(y_test, y_pred)
print('Mean Squared Error of K-Nearest Neighbors Regression is:', mse)
```

```python
plt.hist(y_pred, bins = 'sqrt', histtype = 'step', label = 'Predicted')
plt.hist(y_test, bins = 'sqrt', histtype = 'step', label = 'Actual')
plt.legend(loc = 'upper right')
plt.xlabel('Proporsi Masyarakat Berada di Rumah')
plt.savefig('KNN.png', dpi = 300)

plt.show()
```

## Multiple Linear Regression
